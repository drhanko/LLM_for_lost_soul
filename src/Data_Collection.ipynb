```python
import pandas as pd
from datasets import load_dataset
```

## empathetic_dialogues

```python
empathetic_dialogues = load_dataset("empathetic_dialogues",trust_remote_code=True)
print("1")
print(empathetic_dialogues)
```

```python
empathetic_dialogues
```

```python
rows = []

for item in empathetic_dialogues["train"]:
    rows.append({
        "dataset": "empathetic_dialogues",
        "conversation_id": item["conv_id"],
        "role": "user" if item["speaker_idx"] == 0 else "assistant",
        "text": item["utterance"],
        "emotion": item["context"]
    })

df_empathetic = pd.DataFrame(rows)
df_empathetic.to_csv("./data/empathetic_dialogues_train.csv", index=False)
# df_empathetic_train
df_empathetic
```

```python

rows = []

for item in empathetic_dialogues["validation"]:
    rows.append({
        "dataset": "empathetic_dialogues",
        "conversation_id": item["conv_id"],
        "role": "user" if item["speaker_idx"] == 0 else "assistant",
        "text": item["utterance"],
        "emotion": item["context"]
    })

df_empathetic = pd.DataFrame(rows)
df_empathetic.to_csv("./data/empathetic_dialogues_validation.csv", index=False)
# df_empathetic_val
df_empathetic
```

```python

rows = []

for item in empathetic_dialogues["test"]:
    rows.append({
        "dataset": "empathetic_dialogues",
        "conversation_id": item["conv_id"],
        "role": "user" if item["speaker_idx"] == 0 else "assistant",
        "text": item["utterance"],
        "emotion": item["context"]
    })

df_empathetic = pd.DataFrame(rows)
df_empathetic.to_csv("./data/empathetic_dialogues_test.csv", index=False)
# df_empathetic_val
df_empathetic
```

## GoEmotions

```python
go_emotions = load_dataset("go_emotions",trust_remote_code=True)
print("3")
print(go_emotions)
```
